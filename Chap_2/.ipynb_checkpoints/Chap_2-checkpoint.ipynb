```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(housing_csv, test_size = 0.2, random_state = 42)
```

```python
import numpy as np
housing_csv["income_cat"] = np.ceil(housing_csv["median_income"] / 1.5)
housing_csv["income_cat"].where(housing_csv["income_cat"] < 5, 5.0, inplace= True)
```

```python
plt.hist(housing_csv["income_cat"])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits= 1, test_size= 0.2, random_state=42)
for train_idx , test_idx in split.split(housing_csv, housing_csv["income_cat"]):
    strat_train, strat_test = housing_csv.loc[train_idx], housing_csv.loc[test_idx]
```

```python
housing_csv["income_cat"].value_counts() / len(housing_csv)
```

```python
for set_ in (strat_train, strat_test):
    set_.drop("income_cat", axis= 1, inplace = True)
```

```python
housing_train = strat_train.copy()
```

```python
housing_train.plot(kind = "scatter", x= 'longitude', y= 'latitude')
```

```python
housing_train.plot(kind = "scatter", x= 'longitude', y= 'latitude', alpha = 0.05)
```

```python
housing_train.plot(kind = "scatter", x= 'longitude', y= 'latitude', alpha = 0.4,
                   s = housing_train["population"] / 100,  label = "population", figsize=(10, 7), 
                   c = "median_house_value", cmap = plt.get_cmap("jet"), colorbar = 1, sharex = 0)
plt.legend()
```
